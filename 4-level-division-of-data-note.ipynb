Data can be quantified and qualified, there are 4 types:

1. nominal level(定類)
2. ordinal level(定序)
3. interval level(定距)
4. ratio level(定比)


Nominal level: <br>
Examples: Grade, Lastname, Surname, Animaltype <br>
Math operation available: value_counts() only, cannot add / subtract / division, no mean. <br>
Chart: bar, pie<br>
<br>
Ordinary level: <br>
Data are naturally ordered, can be ranked. <br>
Examples: Exam grades(A-F), Likert scale(1-10) <br>
Math operation: percentile, mean <br>
Chart: stem and leaf plots, box plots <br>
<br>
Interval level: <br>
Data difference are meaningful, we can order and add / subtract it.<br>
Examples: Temperature<br>
Math operation: Add, subtration, mean, std<br>
Chart: histogram, line graph(Value in barchart is not continuous, but in histogram it is continuous and meaningful.)<br>
<br>
Ratio level: <br>
There exist the concept of true zero in data, we can do multiplication / subtraction. <br>
Example: Money value <br>
Temperature have no ratio, it is not meaningful to said A is colder than B for twice. But it is meaninful in money. <br>
Math operation: Multiply, Subtract <br>
Chart: stem and leaf plots, box plots <br>
